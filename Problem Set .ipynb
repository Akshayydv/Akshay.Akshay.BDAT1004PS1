# Question 3

```python
double triangleArea(int a, int b, int c)
{
    float s = (a + b + c) / 2;
    return Math.Sqrt(s * (s - a) * (s - b) * (s - c));
    // Calculated Square Root using Math object method Sqrt()
}

Console.WriteLine(triangleArea(2, 2, 2));
```

# Question 4

```python
Console.Write("Input the no.of elements to be stored in the array :");
int N = int.Parse(Console.ReadLine());
int[] nums = new int[N], oddNums = new int[N], evenNums = new int[N];
int oddCnt = 0, evenCnt = 0;

Console.WriteLine($"Input {N} elments in the array :");
for (int i = 0; i < N; i++)
{
    Console.Write($"element - {i} : ");
    nums[i] = int.Parse(Console.ReadLine());
}

for (int i = 0; i < N; i++)
{
    if (nums[i] % 2 == 0)
        evenNums[evenCnt++] = nums[i];
    else
        oddNums[oddCnt++] = nums[i];
}

Console.WriteLine("The Even elements are:");
for (int i = 0; i < evenCnt; i++)
{
    Console.Write($"{evenNums[i]} ");
}

Console.WriteLine("\nThe Odd elements are:");
for (int i = 0; i < oddCnt; i++)
{
    Console.Write($"{oddNums[i]} ");
}
```

# Question 7

```python
def bldcount(file_name):
  text_file = open(file_name, "r")
  data = text_file.read()
  text_file.close()
  bloodType=data.split()
  a=0
  b=0
  ab=0
  o=0
  oo=0
  for i in bloodType:
    if i=='A':
      a+=1
    elif i=='B':
      b+=1
    elif i=="AB":
      ab+=1
    elif i=='O':
      o+=1
    elif i=="oo":
      oo+=1
  print("There are "+str(a)+" patients of blood type A.")
  print("There are "+str(b)+" patients of blood type B.")
  print("There are "+str(ab)+" patients of blood type AB.")
  print("There are "+str(o)+" patients of blood type O.")
  print("There are "+str(oo)+" patients of blood type OO.")
bldcount("bloodtype1.txt")

```

# Question 8

```python
import codecs

def curconv(curr,amnt):
  pairs = [line.split() for line in codecs.open("currencies.txt",encoding="utf-8")]
  for i in range(len(pairs)):
    if curr == pairs[i][0]:
      break
  return amnt*float(pairs[i][1])
curconv('JPY',100)
```
